```python
##SUPORTING FUNCTIONS FOR THE set_up_game FUNCTION
# 1. draw_train_card: Draws a train card from the top of the deck for game setup. Adds the card to the player's hand and removes it from the deck.
# 2. draw_and_return_ticket: Draws the first three cards from the ticket deck. Adds them to the player's hand. Allows the player to return up to one card to the bottom of the deck, removing it from their hand.

def draw_train_card(player):
  if len(board_activ['train_card_deck']) > 0:
    given_card = board_activ['train_card_deck'][0]
    players[player]["train_cards_in_hand"][given_card] += 1
    board_activ['train_card_deck'].remove(given_card)
    board_activ['train_car_cards'][given_card] -= 1


def draw_and_return_ticket(player, posibel_returns):
  drawn_tickets=[]
  for _ in range(3):
    if len(board_activ["tickets"]) > 0:
      given_card = board_activ["tickets"][0]
      players[player]["tickets_in_hand"].append(given_card)
      drawn_tickets.append(given_card)
      board_activ["tickets"].remove(given_card)
    else:
      break
  #print(f"tickets drawn: {drawn_tickets}")
  for _ in range(random.choice(range(posibel_returns+1))):
    if len(drawn_tickets) > 0:
      chosen_ticket_to_return = random.choice(drawn_tickets)
      drawn_tickets.remove(chosen_ticket_to_return)
      players[player]["tickets_in_hand"].remove(chosen_ticket_to_return)
      board_activ["tickets"].append(chosen_ticket_to_return)
      #print(f"ticket returned {chosen_ticket_to_return}")
    else:
      break
  return drawn_tickets

# SET UP GAME FUNCTION:
# - Randomly chooses the starting player and reorders the player_list.
# - Shuffles and forms the train card deck and the tickets deck.
# - Sets up the player dictionary.
# - Draws an initial 4 train car cards and 3 ticket cards per player. Players can return up to 1 ticket.

def set_up_game(player_count):
  print("GAME SET UP STARTED: CHOSSİNG START PLAYER, FORMİNG DECKS, DRAWİNG COLOR (x4) AND TRAİN CARDS(x3)+RETURN")
  board_activ["player_list"] = []

  # Randomly chooses a starting player and reorders the player list accordingly.
  start_player_choice_int =random.choice(range(1, player_count+1))


  board_activ["player_list"] = [f"player_{i%player_count}" for i in range(start_player_choice_int, start_player_choice_int+player_count)]
  board_activ["player_list"] =list(map(lambda x: f"player_{player_count}" if x == 'player_0' else x, board_activ["player_list"] ))


  # Shuffles and creates the train card deck.
  for color, count in board_activ['train_car_cards'].items():
    board_activ['train_card_deck'].extend([color] * count)
    random.shuffle(board_activ['train_card_deck'])
    #print(f"Train card deck initialized with {len(board_activ['train_card_deck'])} cards.")

  #shuffle the tickets list
  random.shuffle(board_activ["tickets"])

  #set up players dict
  for player in board_activ["player_list"]:
    players[player]= {
        'train_cards_in_hand': {"pink": 0, "white": 0, "blue": 0, "yelow": 0, "orange": 0, "black": 0, "red": 0, "green": 0, "lok": 0},
        'tickets_in_hand': [],
        'completed_tickets': [],
        'possible_connecting_cities': [],
        'score': 0,
        'remaining_trains': 45,
        'player_type': 'AI',
        "routs_claimed_id": []
    }

  # Draws an initial 4 train car cards and 3 ticket cards per player. Players can return up to 1 ticket.
  for player in players:
    for _ in range(4):
      draw_train_card( player=player)
    draw_and_return_ticket(player=player, posibel_returns=1)

  # Prints the results or state of the game after the set up is complete.
  for player in players:
    print(f" {player} has {players[player]['train_cards_in_hand']} color and {players[player]['tickets_in_hand']} tickets in hand" )

```

draw_train_cards_ig FUNCTION

```python
##SUPORTING FUNCTIONS FOR THE draw_train_cards_ig_3 FUNCTION
# 1. reshuffle_and_add_discarded_car_cards: Handles reshuffling and adding discarded car cards back into the deck.
# 2. calculate_pickable_car_cards: Determines available cards, providing lists of face-up cards and the card from the top of the deck.
# 3. pick_card_and_update_decks:
#    - Initializes a reshuffle if fewer than 15 cards remain in the deck.
#    - Discards face-up cards if there are more than two locomotives showing.
#    - Hides locomotive cards from the face-up display if it's the second draw in a turn.
#    - Allows the player to randomly choose between taking a card from the face-up display or drawing from the top of the deck.
#    - Adds the chosen card to the player's hand and removes it from its source deck.

def reshuffle_and_add_dismised_car_cards():
  dismised_car_cards_deck = []
  for color, count in board_activ["discard_pile_train_cards"].items():
    board_activ['train_car_cards'][color]+= count
    dismised_car_cards_deck.extend([color] * count)
  random.shuffle(dismised_car_cards_deck)
  board_activ["train_card_deck"].extend(dismised_car_cards_deck)
  board_activ["discard_pile_train_cards"] = {"pink": 0, "white": 0, "blue": 0, "yelow": 0, "orange": 0, "black": 0, "red": 0, "green": 0, "lok": 0}


def calculate_pickable_car_cards():
  face_up_cards = board_activ['train_card_deck'][:5]
  first_deck_card = board_activ['train_card_deck'][5]
  return face_up_cards, first_deck_card


def pick_card_and_updated_decks(player, noLok=False):
  # Reshuffles and adds discarded cards if the train car card deck has fewer than 15 cards.
  if len(board_activ['train_card_deck']) < 15:
    reshuffle_and_add_dismised_car_cards()
    print("reshuffeld train car card deck")

  # Determines the available face-up cards and the cards at the top of the main deck.
  face_up_cards, first_deck_card = calculate_pickable_car_cards()

  # Removes all face-up cards if there are more than two locomotives.
  if face_up_cards.count("lok") > 2:
    # Moves all cards to the discard pile.
    del board_activ['train_card_deck'][:5]
    for card in face_up_cards:
      board_activ['train_car_cards'][card] -= 1
      board_activ["discard_pile_train_cards"][card]+=1
      board_activ['train_car_cards'][card]-=1
    print("all face up cards dismised becauase more than 2 loks")
    #print(f"new deck : {calculate_pickable_car_cards()}")

  # Recalculates if the deck was altered due to more than two locomotives being face-up.
  face_up_cards, first_deck_card = calculate_pickable_car_cards()

  deck_choice=random.choice(["face_up_cards", "first_deck_card"])

  # When 'noLok' is set to true, clear all locomotive cards from the visible face-up cards.
  if noLok==True:
    face_up_cards = [i for i in face_up_cards if i != "lok"]

  # Picks a card based on the player's choice and updates the deck after the selection.
  if deck_choice == "face_up_cards":
    choosen_car_card = random.choice(face_up_cards)
  if deck_choice == "first_deck_card":
    choosen_car_card=first_deck_card

  # Adds the card to the player's hand, removes it from the deck list, and updates the tracking dictionary.
  players[player]["train_cards_in_hand"][choosen_car_card] += 1
  board_activ['train_card_deck'].remove(choosen_car_card)
  board_activ['train_car_cards'][choosen_car_card] -= 1

  return choosen_car_card, deck_choice


# DRAW TRAIN CAR CARD FUNCTION:
def draw_train_cards_ig_3(player):
  # First draw:
  picked_cards = []
  choosen_car_card_1, first_choice = pick_card_and_updated_decks(player=player, )
  picked_cards.append(choosen_car_card_1)

  # Second draw:
  if first_choice == "face_up_cards" and choosen_car_card_1  =="lok":
    print("no more cards to pick because lok was pikt")

  else:
    choosen_car_card_2, deck_choice = pick_card_and_updated_decks(player=player, noLok=True)
    picked_cards.append(choosen_car_card_2)

  print(f"{player} cards choosen: {picked_cards}")
```

claim_route FUNCTION

```python
#Route Claiming Function: Current Approach and Potential Enhancements

#For the route claiming function, I've decided to implement only the essential game constraints:
#   1. The route must not be owned by another player.
#   2. The player must have enough wagons left to build the route.
#   3. The player must have enough color cards and additional locomotive cards available.

#Should a change in this approach be desired, the following basic logic enhancements could be considered:
#   1. Remove secondary routes: Eliminate routes that are secondary to routes a player has already claimed.
#   2. Limit claimable routes to ticket connections: Restrict claimable routes only to those connected to the start and end cities of an active ticket.
#   3. Connection logic for claimed routes: Add logic to connect two cities even if routes in other directions are already claimed.
#   4. Prioritize longer routes: Implement logic to favor claiming longer routes.
#   5. noLok usage logic: Integrate specific logic for when to use noLok (True or False).


##SUPORTING FUNCTIONS FOR THE claim_route FUNCTION
# check_if_route_build_possible:
#    Loops through all possible routes to determine if:
#    - The route is currently unowned.
#    - The player has enough wagons available.
#    - The player possesses the required train car cards and, if necessary, locomotive cards.

length_to_score={1:1,
                 2:3,
                 3:4,
                 4:7,
                 5:10,
                 6:15,}

def check_if_route_build_possible(player,  routs_data, no_lok=True):
  buildable_routes = []
  player_hand=players[player]["train_cards_in_hand"]
  all_routs_id = board_static["routes"].keys()

  for route_as_id in all_routs_id:

    owner_status = routs_data[route_as_id]["owner"]
    required_length = routs_data[route_as_id]['length']
    route_color = routs_data[route_as_id]['color']
    available_locomotives = player_hand.get('lok', 0)

    # owner is none?
    if owner_status is not None:
      #print(f" {route_as_id} already claimed")
      continue

    # enougth wagons left?
    if players[player]["remaining_trains"]< required_length:
      #print(f" for  {route_as_id} not enougth wagons {player_d[player]['remaining_trains']}/{required_length}")
      continue

    # enougth train cards? choose randomly if using loks is wanted or not?
    if no_lok:
      available_locomotives = 0

    if route_color != 'grey':
        available_matching_color_cards = player_hand.get(route_color, 0)
        total_available = available_matching_color_cards + available_locomotives

        if total_available >= required_length:
          buildable_routes.append((route_as_id, route_color, required_length))
          #print(f"{route_as_id} appended to buildabable routs")
    else:
      for color_option in board_static["standard_colors"]:
        available_specific_color_cards = player_hand.get(color_option, 0)
        total_available_for_this_option = available_specific_color_cards + available_locomotives
        if total_available_for_this_option >= required_length:
          buildable_routes.append((route_as_id, color_option, required_length))
          #print(f"{route_as_id} appended to buildabable routs")

  return buildable_routes


# CLAİM ROUTE FUNCTION:
# Takes 'route_to_claim_tuple' as input.
# Removes required train car cards from the player's hand and adds them to the discard pile.
# Changes the status of the claimed route to owned by the player.
# Adds points to the player's score and removes wagons from the player's supply.

def claim_route(player, rout_to_claim_touple, routs_data):

  player_hand=players[player]["train_cards_in_hand"]
  route_id = rout_to_claim_touple[0]
  color = rout_to_claim_touple[1]
  length = rout_to_claim_touple[2]

  # Remove train car cards from the player's hand.
  # Add train car cards to the discard pile.
  for _ in range(length):
    if player_hand[color]>0:
      player_hand[color] -= 1
      board_activ["discard_pile_train_cards"][color] += 1
    else:
      player_hand["lok"] -= 1
      board_activ["discard_pile_train_cards"]["lok"] += 1

  # Changes the ownership status of the claimed route.
  routs_data[route_id]["owner"] = player
  # Adds the route ID to the list of claimed routes (routs_claimd_id).
  players[player]["routs_claimed_id"].append(route_id)
  # Add points to the player's score.
  players[player]["score"] += length_to_score[length]
  # Decrements the number of wagons available to the player.
  players[player]['remaining_trains'] -= length
  board_activ["claimed_routes"].append((route_id, player))

```

check_ticket_fulfilled FUNCTION

```python
# Creates a player network (graph).
# Loops through the tickets in the player's hand to determine if each given ticket is fulfilled using Breadth-First Search (BFS).
def check_ticket_fulfilled(player, ticket):

  start_city = ticket['city_1_id']
  end_city = ticket['city_2_id']

  # Loops through all routes the player has claimed to create their network, for example: {'STF': {'DEN', 'ELP'}, 'ELP': {'STF'}, 'DEN': {'STF'}}.
  player_network = {}
  player_claimed_route_ids = players[player]['routs_claimed_id']
  for route_id in player_claimed_route_ids:
    route_details = board_static["routes"][route_id]
    city1 = route_details['city_1_id']
    city2 = route_details['city_2_id']
    if city1 not in player_network:
      player_network[city1] = set()
    if city2 not in player_network:
      player_network[city2] = set()
    player_network[city1].add(city2)
    player_network[city2].add(city1)


  # Skips checking the ticket if either the start or end city is not present in the player's claimed network (route not connected).
  if start_city not in player_network or end_city not in player_network:
    return False

  # Executes a Breadth-First Search (BFS) algorithm to determine if a path exists.
  queue = deque([start_city])
  visited = {start_city}
  while queue:
    current_city = queue.popleft()

    if current_city == end_city:
      print("ticket completed, adding score, removing ticket from hand, adding ticket to completed tickets")
     # Adds the ticket to 'completed_tickets' and removes it from the player's hand.
      players[player]['tickets_in_hand'].remove(ticket)
      players[player]['completed_tickets'].append(ticket)
      # Add score to the player's total.
      players[player]['score'] += ticket['points']
      return True # Path found!
    for neighbor in player_network.get(current_city, set()):
      if neighbor not in visited:
        visited.add(neighbor)
        queue.append(neighbor)
  #print("No path found")

  return False
```

in_game_player_actions FUNCTION

```python
# PLAYER ACTION FUNCTION:
# Allows the player to choose one of three possible actions:
# - Claim a route
# - Draw a destination ticket
# - Draw a train car card

def in_game_player_actions(player):
    print(f"TURN {board_activ['turn_count']}")
    buildable_routes_noLok_choice = random.choice([True, False])
    players[player]["buildable_routes"]=check_if_route_build_possible(player=player, routs_data=board_static['routes'], no_lok=buildable_routes_noLok_choice)

# IMPORTANT: Some of these checks might be redundant with lower-level logic. Please review and synchronize the approaches if needed.

  # Relevant checks for player actions:
  # claim_route_possible = players[player]["buildable_routes"] # Checks if the player has any routes they can build.
  # draw_ticket = len(board_activ["tickets"]) > 1             # Checks if there's more than one ticket available to draw.
  # draw_train_cards = len(board_activ["train_card_deck"]) > 1 # Checks if there's more than one train card available to draw.

    possible_actions=[]
    if players[player]["buildable_routes"]:
      possible_actions.append('claim_route')
    if len(board_activ["tickets"])>1:
      possible_actions.append('draw_ticket')
    if len(board_activ["train_card_deck"])>2:
      possible_actions.append('draw_train_cards')

    if possible_actions:
      chosen_action = random.choice(possible_actions)

    print(f"{player} chosses to: {chosen_action}")


# CLAIM ROUTE
    if chosen_action == 'claim_route':
      # The AI randomly picks one of the available routes from the 'buildable_routes' list to claim.
      rout_to_claim_touple=random.choice(players[player]["buildable_routes"])
      claim_route(player=player, rout_to_claim_touple=rout_to_claim_touple, routs_data=board_static['routes'])
      print(f"rout claimed {rout_to_claim_touple}")

      # Verify if destination tickets have been successfully fulfilled.
      for ticket in players[player]['tickets_in_hand']:
        check_ticket_fulfilled(player=player, ticket=ticket)


# DRAW TICKET
    elif chosen_action == 'draw_ticket':
      drawn_tickets = draw_and_return_ticket(player=player, posibel_returns=2)
      print(f"Drawn Tickets {drawn_tickets}")

# DRAW TRAIN CAR CARD
    elif chosen_action == 'draw_train_cards':
      draw_train_cards_ig_3( player=player)


    else:
      # This case should ideally not happen in Ticket to Ride unless the game is over
      # or in a very specific, impossible state.
      print(f"Warning: No possible actions for {player}. Game state might be stuck.")





```

plot_map FUNCTION

```python
# Create a dictionary to quickly get coordinates by city ID
city_coords_map = {}
for city in board_static["cities"]:
    x_str, y_str = city["coo"].split('-')
    city_coords_map[city["id"]] = (int(x_str), int(y_str))

color_map = {
    'pink': '#FF69B4',   # Hot pink
    'white': '#D3D3D3',  # Light gray
    'blue': '#4682B4',   # Steel Blue
    'yelow': '#FFD700',  # Gold
    'orange': '#FFA500', # Orange
    'black': '#000000',  # Black
    'red': '#DC143C',    # Crimson
    'green': '#228B22',  # Forest Green
    'grey': '#A9A9A9',   # Dark Gray for 'grey' routes
}


# Prepare Data for Plotting Cities
x_cities = []
y_cities = []
city_names = []
for city_id, (x, y) in city_coords_map.items():
    x_cities.append(x)
    y_cities.append(y)
    # Find the city name by ID for labels
    name = next((c['CName'] for c in board_static["cities"] if c['id'] == city_id), city_id)
    city_names.append(name)


def plot_ttr_map_2(ax, cities_data, routes_data, claimed_routes={}):
    ax.clear()


    city_coords_map = {}
    city_names = []
    x_cities = []
    y_cities = []

    for city in cities_data:
        x_str, y_str = city["coo"].split('-')
        x, y = int(x_str), int(y_str)
        city_coords_map[city["id"]] = (x, y)
        city_names.append(city["id"])
        x_cities.append(x)
        y_cities.append(y)



    connection_counts = {}
    offset_amount = 0.15

    for route_id, route_info in routes_data.items():
        city1_id = route_info['city_1_id']
        city2_id = route_info['city_2_id']

        if city1_id not in city_coords_map or city2_id not in city_coords_map:
            print(f"Warning: Route {route_id} connects to unknown city(s). Skipping.")
            continue

        x1, y1 = city_coords_map[city1_id]
        x2, y2 = city_coords_map[city2_id]

        connection_key = tuple(sorted((city1_id, city2_id)))
        connection_counts[connection_key] = connection_counts.get(connection_key, 0) + 1
        current_connection_idx = connection_counts[connection_key]

        offset_x, offset_y = 0, 0
        if current_connection_idx > 1:
            vec_x, vec_y = x2 - x1, y2 - y1
            perp_vec_x, perp_vec_y = -vec_y, vec_x
            magnitude = np.sqrt(perp_vec_x**2 + perp_vec_y**2)
            if magnitude > 0:
                perp_vec_x /= magnitude
                perp_vec_y /= magnitude

            offset_direction = 1 if current_connection_idx % 2 != 0 else -1
            offset_factor = (current_connection_idx // 2) * offset_amount
            offset_x = offset_direction * offset_factor * perp_vec_x
            offset_y = offset_direction * offset_factor * perp_vec_y

        route_color_name = route_info['color']
        line_color = color_map.get(route_color_name, 'gray')
        line_width = 2
        linestyle='--'
        alpha=0.3

        # Override color if route is claimed
        if route_id in claimed_routes and claimed_routes[route_id] is not None:
            owner_color = claimed_routes[route_id]
            line_color = owner_color
            line_width = 5
            linestyle='-'
            alpha=1

        ax.plot(
            [x1 + offset_x, x2 + offset_x],
            [y1 + offset_y, y2 + offset_y],
            color=line_color,
            linewidth=line_width,
            linestyle=linestyle,
            alpha=alpha,
            zorder=1
        )

    ax.scatter(x_cities, y_cities, color='red', s=150, zorder=5) # Larger points for visibility

    # Add labels for each city
    for i, city_name in enumerate(city_names):
        ax.text(x_cities[i], y_cities[i] + 0.3, city_name, fontsize=13, ha='center', va='center', zorder=6, fontweight="bold")

    ax.set_xlim(0, 23)
    ax.set_ylim(0, 14)

    ax.set_xlabel("X Coordinate")
    ax.set_ylabel("Y Coordinate")
    ax.set_title("Ticket to Ride Map - Cities and Routes")
    ax.grid(True, linestyle='--', alpha=0.5, zorder=0)

    ax.set_xticks(range(-1, 24))
    ax.set_yticks(range(-1, 14))

    ax.set_aspect('equal', adjustable='box')


def plot_map(ax):
  board_activ["claimed_routes"]
  routes_owner_dict = {item[0]: item[1] for item in board_activ["claimed_routes"]}

  player_color_map = {
      'player_1': 'red',
      'player_2': 'blue',
      'player_3': 'green',
      'player_4': 'yellow'
  }


  colored_routes_dict = {}
  for route_id, owner_id in routes_owner_dict.items():
    if owner_id in player_color_map:
      colored_routes_dict[route_id] = player_color_map[owner_id]

  plot_ttr_map_2(ax, board_static["cities"], board_static["routes"], claimed_routes=colored_routes_dict)
```

## MAIN GAME LOOP

```python
# Set up game
players={}
set_up_game(player_count=3)

# Anchor map
fig, ax = plt.subplots(figsize=(15, 10))

# Sets the game state and opens the main game loop.
game_on=True
last_round_triggered_by=None
while game_on:
  for player in players:
    board_activ['current_player'] = player

    in_game_player_actions(player)


    clear_output(wait=True)
    plot_map(ax)
    display(fig)
    time.sleep(0.1)

    # Checks for the game end trigger.
    if players[player]['remaining_trains']<= 2:
      last_round_triggered_by=player
      game_on=False

      print(f"last round trigered by {player}")
      break
    board_activ["turn_count"] += 1

# Reshuffles the player list to determine the player order  for the final round.
player_deque = deque(board_activ["player_list"])
trigger_index = board_activ["player_list"].index(last_round_triggered_by)
player_deque.rotate(-(trigger_index + 1))
print(f"last round started with folowing order{player_deque} as trigered by {last_round_triggered_by}")

# Plays the final round and calculates the final scores.
for player in player_deque:
  in_game_player_actions(player)
  for ticket in players[player]['tickets_in_hand']:
    players[player]['score']-=ticket["points"]
    print(f"{ticket} not completed: {ticket['points']} subtracted from total oints")


plot_map(ax)
for player in players:
  print(f"------------------{player}-----------------\n  Points: {players[player]['score']}\n  Completet tickets: {players[player]['completed_tickets']}\n  Wagons remainig: {players[player]['remaining_trains']}\n  Total tickets drawn: {len(players[player]['tickets_in_hand'])}" )

```

## DT  (Decision Together) MAIN GAME LOOP
For this in-game player action function, a design decision was made to consolidate all random choices within a single, higher-level function. This approach ensures that decision-making logic isn't dispersed throughout various lower-level functions, improving maintainability and clarity. As a result, certain foundational functions required modification and have been renamed with a _DT suffix to indicate their deterministic nature under this new structure.

```python
def reshuffle_and_add_dismised_car_cards():
  dismised_car_cards_deck = []
  for color, count in board_activ["discard_pile_train_cards"].items():
    board_activ['train_car_cards'][color]+= count
    dismised_car_cards_deck.extend([color] * count)
  random.shuffle(dismised_car_cards_deck)
  board_activ["train_card_deck"].extend(dismised_car_cards_deck)
  board_activ["discard_pile_train_cards"] = {"pink": 0, "white": 0, "blue": 0, "yelow": 0, "orange": 0, "black": 0, "red": 0, "green": 0, "lok": 0}




def updated_decks_DT(noLok=False):
  # Reshuffles and adds discarded cards if the train car card deck has fewer than 15 cards.
  if len(board_activ['train_card_deck']) < 15:
    reshuffle_and_add_dismised_car_cards()
    print("reshuffeld train car card deck")

  # Determines the available face-up cards and the cards at the top of the main deck.
  face_up_cards = board_activ['train_card_deck'][:5]
  first_deck_card = board_activ['train_card_deck'][5]

  # Removes all face-up cards if there are more than two locomotives.
  if face_up_cards.count("lok") > 2:
    # Moves all cards to the discard pile.
    del board_activ['train_card_deck'][:5]
    for card in face_up_cards:
      board_activ['train_car_cards'][card] -= 1
      board_activ["discard_pile_train_cards"][card]+=1


  # Recalculates if the deck was altered due to more than two locomotives being face-up.
  face_up_cards = board_activ['train_card_deck'][:5]
  first_deck_card = board_activ['train_card_deck'][5]

  # When 'noLok' is set to true, clear all locomotive cards from the visible face-up cards.
  if noLok==True:
    face_up_cards = [i for i in face_up_cards if i != "lok"]

  return face_up_cards, first_deck_card

def add_card_to_player_hand_DT(choosen_car_card):
  # Adds the card to the player's hand, removes it from the deck, and updates the game's tracking data.
  players[player]["train_cards_in_hand"][choosen_car_card] += 1
  board_activ['train_card_deck'].remove(choosen_car_card)
  board_activ['train_car_cards'][choosen_car_card] -= 1


```

```python
def draw_ticket_DT(player,):
  drawn_tickets=[]
  for _ in range(3):
    if len(board_activ["tickets"]) > 0:
      given_card = board_activ["tickets"][0]
      players[player]["tickets_in_hand"].append(given_card)
      drawn_tickets.append(given_card)
      board_activ["tickets"].remove(given_card)
    else:
      break
  return drawn_tickets
  #print(f"tickets drawn: {drawn_tickets}")

def return_ticket_DT(tickets_to_return, player):
  for ticket in tickets_to_return:

    players[player]["tickets_in_hand"].remove(ticket)
    board_activ["tickets"].append(ticket)
```

```python
def set_up_game_DT(player_count):
  print("GAME SET UP STARTED: CHOSSİNG START PLAYER, FORMİNG DECKS, DRAWİNG COLOR (x4) AND TRAİN CARDS(x3)+RETURN")
  board_activ["player_list"] = []

  # Randomly chooses a starting player and reorders the player list accordingly.
  start_player_choice_int =random.choice(range(1, player_count+1))


  board_activ["player_list"] = [f"player_{i%player_count}" for i in range(start_player_choice_int, start_player_choice_int+player_count)]
  board_activ["player_list"] =list(map(lambda x: f"player_{player_count}" if x == 'player_0' else x, board_activ["player_list"] ))
  #print(board_activ["player_list"])

  # Shuffles and creates the train card deck.
  for color, count in board_activ['train_car_cards'].items():
    board_activ['train_card_deck'].extend([color] * count)
    random.shuffle(board_activ['train_card_deck'])
    #print(f"Train card deck initialized with {len(board_activ['train_card_deck'])} cards.")

  #shuffle the tickets list.
  random.shuffle(board_activ["tickets"])

  #set up players dict
  for player in board_activ["player_list"]:
    players[player]= {
        'train_cards_in_hand': {"pink": 0, "white": 0, "blue": 0, "yelow": 0, "orange": 0, "black": 0, "red": 0, "green": 0, "lok": 0},
        'tickets_in_hand': [],
        'completed_tickets': [],
        'possible_connecting_cities': [],
        'score': 0,
        'remaining_trains': 45,
        'player_type': 'AI',
        "routs_claimed_id": []
    }

  #print(players)

# Draws an initial 4 train car cards.
  for player in players:
    for _ in range(4):
      draw_train_card(player=player)

  for player in players:
    print(f" {player} has {players[player]['train_cards_in_hand']} color and {players[player]['tickets_in_hand']} tickets in hand" )
```

```python
# PLAYER ACTION FUNCTION:
# Allows the player to choose one of three possible actions:
# - Claim a route
# - Draw a destination ticket
# - Draw a train car card


def in_game_player_actions_DT(player):
  print(f"TURN {board_activ['turn_count']}")

  # For the first four turns, players only select their starting tickets and can return up to one of their chosen tickets.
  if board_activ['turn_count'] < len(players):
    drawn_tickets = draw_ticket_DT(player=player)
    print(f"{player} draw {drawn_tickets}")
    num_to_return=random.choice([0, 1])
    if num_to_return > 0:
      tickets_to_return = random.sample(drawn_tickets, num_to_return)
      return_ticket_DT(tickets_to_return=tickets_to_return, player=player)
      print(f"tickets returned {tickets_to_return}")


  else:
    buildable_routes_noLok_choice = random.choice([True, False])
    players[player]["buildable_routes"]=check_if_route_build_possible(player=player, routs_data=board_static['routes'], no_lok=buildable_routes_noLok_choice)

    # Relevant checks for player actions:
    # claim_route_possible = players[player]["buildable_routes"] # Checks if the player has any routes they can build.
    # draw_ticket = len(board_activ["tickets"]) > 1             # Checks if there's more than one ticket available to draw.
    # draw_train_cards = len(board_activ["train_card_deck"]) > 1 # Checks if there's more than one train card available to draw.


    possible_actions=[]
    if players[player]["buildable_routes"]:
      possible_actions.append('claim_route')
    if len(board_activ["tickets"])>1:
      possible_actions.append('draw_ticket')
    if len(board_activ["train_card_deck"])>2:
      possible_actions.append('draw_train_cards')

    if possible_actions:
      chosen_action = random.choice(possible_actions)

    #print(f"-----------------------------------{player} Turn---------------------------------------------")
    #print(f"  Possible actions: {possible_actions}")
    print(f"{player} chosses to: {chosen_action}")

#CLAİME ROUT
    if chosen_action == 'claim_route':
      # AI randomly picks one of the available routes to claim
      rout_to_claim_touple=random.choice(players[player]["buildable_routes"])
      claim_route(player=player, rout_to_claim_touple=rout_to_claim_touple, routs_data=board_static['routes'])
      print(f"rout claimed {rout_to_claim_touple}")
      for ticket in players[player]['tickets_in_hand']:
        check_ticket_fulfilled(player=player, ticket=ticket)

      #print("fullfilled ticket check run throuth")


# DRAW TİCKET
    elif chosen_action == 'draw_ticket':
      drawn_tickets = draw_ticket_DT(player=player)
      print(f"{player} draw {drawn_tickets}")
      num_to_return=random.choice([0, 1, 2])
      if num_to_return > 0:
        tickets_to_return = random.sample(drawn_tickets, num_to_return)
        return_ticket_DT( tickets_to_return=tickets_to_return, player= player)
        print(f"{player} returned {tickets_to_return}")



# DRAW TRAİN CAR CARD
    elif chosen_action == 'draw_train_cards':

      #print(f"deck at start of loop: {calculate_pickable_car_cards()}")

      picked_cards = []
      face_up_cards, first_deck_card = updated_decks_DT()

      #FİRST DRAW
      #make first deck choice
      first_choice=random.choice(["face_up_cards", "first_deck_card"])
      #take card
      if first_choice == "face_up_cards":
        choosen_car_card_1 = random.choice(face_up_cards)
      if first_choice == "first_deck_card":
        choosen_car_card_1=first_deck_card
      #add card to player hand
      add_card_to_player_hand_DT(choosen_car_card=choosen_car_card_1)
      picked_cards.append(choosen_car_card_1)


      #check if second draw allowed
      if first_choice == "face_up_cards" and choosen_car_card_1  =="lok":
        print("no more cards to pick because lok was pikt")

      # contiune second draw:
      else:
        face_up_cards, first_deck_card = updated_decks_DT(noLok=True)
        second_choice=random.choice(["face_up_cards", "first_deck_card"])
        if second_choice == "face_up_cards":
          choosen_car_card_2 = random.choice(face_up_cards)
        if second_choice == "first_deck_card":
          choosen_car_card_2=first_deck_card
        add_card_to_player_hand_DT(choosen_car_card=choosen_car_card_2)
        picked_cards.append(choosen_car_card_2)

      #print(f"deck after full loop: {calculate_pickable_car_cards()}")
      print(f"{player} cards choosen: {picked_cards}")



    else:
        # This case should ideally not happen in Ticket to Ride unless the game is over
        # or in a very specific, impossible state.
      print(f"Warning: No possible actions for {player}. Game state might be stuck.")
```
