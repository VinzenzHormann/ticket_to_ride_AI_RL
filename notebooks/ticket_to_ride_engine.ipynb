```python
##SUPORTING FUNCTIONS FIR THE set_up_game FUNCTION
def draw_train_card(stapel, stat_count, player, player_d):
  if len(stapel) > 0:
    given_card = stapel[0]
    player_d[player]["train_cards_in_hand"][given_card] += 1
    stapel.remove(given_card)
    stat_count[given_card] -= 1

def draw_ticket_card(stapel, player, player_d):
  if len(stapel) > 0:
    given_card = stapel[0]
    player_d[player]["tickets_in_hand"].append(given_card)
    stapel.remove(given_card)

def return_ticket_card(stapel, player, player_d, posibel_returns):
  for _ in range(random.choice(range(posibel_returns+1))):
    chosen_ticket_to_return = random.choice(player_d[player]["tickets_in_hand"])
    player_d[player]["tickets_in_hand"].remove(chosen_ticket_to_return)
    stapel.append(chosen_ticket_to_return)
    print(f"ticket returned {chosen_ticket_to_return}")
```

```python
def set_up_game(player_count):
  board_activ["player_count"]= player_count

  # use train_car_cards to shuffel and form the train_card_deck
  for color, count in board_activ['train_car_cards'].items():
    board_activ['train_card_deck'].extend([color] * count)
    random.shuffle(board_activ['train_card_deck'])
    #print(f"Train card deck initialized with {len(board_activ['train_card_deck'])} cards.")

  #shuffle the tickets list.
  random.shuffle(board_activ["tickets"])

  #set up players dict
  for n in range(1, player_count+1):
    players[f"player_{n}"]= {
        'train_cards_in_hand': {"pink": 0, "white": 0, "blue": 0, "yelow": 0, "orange": 0, "black": 0, "red": 0, "green": 0, "lok": 0},
        'tickets_in_hand': [],
        'completed_tickets': [],
        'score': 0,
        'remaining_trains': 45,
        'player_type': 'AI'
    }

  #print(players)

  # draw initial 4 car cards and 3 ticket cards per player. Return up to 1 ticket
  for player in players:
    print("---------------------------------------")
    print(player)
    for _ in range(4):
      draw_train_card(stapel=board_activ['train_card_deck'], stat_count=board_activ['train_car_cards'], player=player, player_d=players)
    print(f"Players card after initial 4 train card draw: {players[player]['train_cards_in_hand']}")
    for _ in range(3):
      draw_ticket_card(stapel=board_activ["tickets"], player=player, player_d=players)
    print(f"Players card after initial 3 ticket draw: {players[player]['tickets_in_hand']}")
    return_ticket_card(stapel=board_activ["tickets"], player=player, player_d=players, posibel_returns=1)
    #print(f"Players card after initial (random) ticket return: {players[player]['tickets_in_hand']}")




```

```python
def plot_ttr_map(cities_data, routes_data, claimed_routes={}):
    """
    Plots the Ticket to Ride map with cities and routes.
    Args:
        cities_data (list): List of city dictionaries.
        routes_data (dict): Dictionary of route dictionaries.
        claimed_routes (dict): Optional dictionary mapping route_id to owner_color
                                for highlighting claimed routes.
    """
    plt.figure(figsize=(15, 10)) # Adjust figure size for better readability

    # Create city coordinates map: ıd to cordinates touple
    city_coords_map = {}
    for city in cities_data:
        x_str, y_str = city["coo"].split('-')
        city_coords_map[city["id"]] = (int(x_str), int(y_str))

    # --- Plot Routes ---
    # To handle double routes, we'll track how many times we've drawn a connection between two cities
    connection_counts = {}
    offset_amount = 0.15 # Small offset for parallel routes


    for route_id, route_info in routes_data.items():
        city1_id = route_info['city_1_id']
        city2_id = route_info['city_2_id']

        # Ensure both cities exist in the map
        if city1_id not in city_coords_map or city2_id not in city_coords_map:
            print(f"Warning: Route {route_id} connects to unknown city(s). Skipping.")
            continue
        # get x and y for both cities from the city_coords_map dict
        x1, y1 = city_coords_map[city1_id]
        x2, y2 = city_coords_map[city2_id]

        # Determine the key for this connection (order-independent)
        connection_key = tuple(sorted((city1_id, city2_id)))

        # Increment connection count
        connection_counts[connection_key] = connection_counts.get(connection_key, 0) + 1
        current_connection_idx = connection_counts[connection_key]

        # Calculate offset for double/multiple routes
        offset_x, offset_y = 0, 0
        if current_connection_idx > 1:
            # Calculate the vector from city1 to city2
            vec_x, vec_y = x2 - x1, y2 - y1
            # Calculate the perpendicular vector (normalized)
            perp_vec_x, perp_vec_y = -vec_y, vec_x
            magnitude = np.sqrt(perp_vec_x**2 + perp_vec_y**2)
            if magnitude > 0:
                perp_vec_x /= magnitude
                perp_vec_y /= magnitude

            # Apply offset: + for second route, - for third, etc.
            offset_direction = 1 if current_connection_idx % 2 != 0 else -1
            offset_factor = (current_connection_idx // 2) * offset_amount
            offset_x = offset_direction * offset_factor * perp_vec_x
            offset_y = offset_direction * offset_factor * perp_vec_y

        # Determine line color an styl
        route_color_name = route_info['color']
        line_color = color_map.get(route_color_name, 'gray') # Default to gray if color not found
        line_width = 5 # Default line width
        linestyle='--'
        alpha=0.6


        # Override color if route is claimed
        if route_id in claimed_routes and claimed_routes[route_id] is not None:
            owner_color = claimed_routes[route_id] # Assuming owner_color is a valid matplotlib color
            line_color = owner_color # Use the player's color for claimed routes
            line_width = 5 # Make claimed routes thicker
            linestyle='-'
            alpha=1 # Make claimed routes more visible

        # Plot the line with offset
        plt.plot(
            [x1 + offset_x, x2 + offset_x],
            [y1 + offset_y, y2 + offset_y],
            color=line_color,
            linewidth=line_width,
            linestyle=linestyle,
            alpha=alpha,
            zorder=1 # Draw routes behind cities
        )


        # Optionally add route length text (can be cluttered on busy maps)
        # mid_x = (x1 + x2) / 2 + offset_x
        # mid_y = (y1 + y2) / 2 + offset_y
        # plt.text(mid_x, mid_y, str(route_info['length']), fontsize=8, color='black', ha='center', va='center', bbox=dict(facecolor='white', alpha=0.7, edgecolor='none', boxstyle='round,pad=0.2'), zorder=2)


    # --- Plot Cities (on top of routes) ---
    plt.scatter(x_cities, y_cities, color='red', s=150, zorder=5) # Larger points for visibility

    # Add labels for each city
    for i, city_name in enumerate(city_names):
        plt.text(x_cities[i], y_cities[i] + 0.3, city_name, fontsize=13, ha='center', va='center', zorder=6, fontweight="bold")

    # --- Set plot limits and labels ---
    plt.xlim(0, 23)
    plt.ylim(0, 14)

    plt.xlabel("X Coordinate")
    plt.ylabel("Y Coordinate")
    plt.title("Ticket to Ride Map - Cities and Routes")
    #plt.grid(True, linestyle='--', alpha=0.5, zorder=0) # Grid behind everything

    plt.xticks(range(-1, 24))
    plt.yticks(range(-1, 14))

    plt.gca().set_aspect('equal', adjustable='box')
    plt.tight_layout()
    plt.show()
```
